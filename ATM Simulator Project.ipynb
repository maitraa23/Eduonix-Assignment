```python
class User:
  def __init__(self,name,pin):
    self.name=name
    self.pin=pin
```

```python
user=User('Arshi',1234)
```

```python
def atm_simulator(user):
  balance=10000
  print('Welcome to ATM Simulator, please enter your name and pin')
  name= input('Please enter your name')
  pin=int(input('Please enter your PIN'))
  if name==user.name and pin==user.pin:
    print(f'Hi {user.name} please select from the below options')
    while True:
      print('1.Balance')
      print('2.Withdraw')
      print('3.Deposit')
      print('4.Exit')
      choice=int(input("Enter your choice:"))
      if choice==1:
        print(f"Your current balance is {balance}")
      elif choice==2:
        withdraw=int(input('Please enter the withdrawal amount:'))
        if withdraw>balance:
          print('Insufficient Fund!')
        else:
          balance=balance-withdraw
          print('Do you wish to see your available balance: 1. Yes. 2. No')
          available_balance=int((input('Enter your choice:')))
          if available_balance==1:
            print(f'Your Available Balance is {balance}')
          else:
              continue
      elif choice==3:
        deposit=int(input('Enter the deposit amount:'))
        print('Amount succesfully deposited, Thank You!')
        balance=balance+deposit
        print('Do you wish to see your available balance: 1. Yes. 2. No')
        available_balance=int((input('Enter your choice:')))
        if available_balance==1:
         print(f'Your Available Balance is {balance}')
        else:
          continue
      elif choice==4:
        print('Thank you!')
        return

  else:
    ('You have entered incorrect details')





atm_simulator(user)
```
